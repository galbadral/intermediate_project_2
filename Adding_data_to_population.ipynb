```python
import pandas as pd
```

```python
df = pd.read_csv("population.csv")
```

```python
df
```

```python
for i in (range(len(df))):
    dfc = pd.read_csv(r'C:\Users\Galbadral\Downloads\Intermediate python proejct 2\countries_frate\\'+str(df['Country Code'][i])+'.csv')
    dfc['inc']=0
    dfc['per']=0
    for x in range(1,len(dfc)):
        dfc['inc'][x]=dfc['fertility'][x]-dfc['fertility'][x-1]
        dfc['per'][x]=(dfc['inc'][x]/dfc['fertility'][x-1])*100
    sum_per=dfc['per'].sum()
    avg_per=sum_per/59
    df['mean_yearly_fertility_change_percentage'][i]=avg_per
    dfc_10=dfc.iloc[49:60]
    dfc_20=dfc.iloc[39:60]
    avg_per_10=dfc_10['per'].sum()/10
    avg_per_20=dfc_20['per'].sum()/20
    df['mean_fertility_10years'][i]=avg_per_10
    df['mean_fertility_20years'][i]=avg_per_20
```

```python
df['yearly_decrease_to_zero_per']=0
```

```python
import numpy
import math
```

```python
for i in range(len(df)):
    df['yearly_decrease_to_zero_per'][i]=df['Population'][i]/df['Lifespan'][i]
    df['yearly_decrease_to_zero_per'][i]=math.trunc(numpy.nan_to_num(df['yearly_decrease_to_zero_per'][i]))
    
```
